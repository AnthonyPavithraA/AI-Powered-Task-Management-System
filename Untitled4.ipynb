```python
# AI-Powered Task Management System

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# NLP libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re

# ML libraries
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder, StandardScaler
import xgboost as xgb

# Download NLTK data
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')
```

```python
print("Loading and exploring the dataset...")

# Loading the dataset
df = pd.read_csv('Data.csv')

print(f"Dataset shape: {df.shape}")
print("\nDataset Info:")
print(df.info())

print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Statistics:")
print(df.describe())

# Checking for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
print("\n=== NLP PREPROCESSING ===")

class NLPPreprocessor:
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
    
    def preprocess_text(self, text):
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters and digits
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenization
        tokens = word_tokenize(text)
        
        # Remove stopwords only, no stemming
        processed_tokens = [
            token
            for token in tokens
            if token not in self.stop_words and len(token) > 2
        ]
        
        return ' '.join(processed_tokens)

# Initialize preprocessor
nlp_processor = NLPPreprocessor()

print("Preprocessing task descriptions...")
df['processed_description'] = df['task_description'].apply(nlp_processor.preprocess_text)

print("Sample processed descriptions:")
for i in range(len(df)):
    print(f"Original: {df['task_description'].iloc[i]}")
    print(f"Processed: {df['processed_description'].iloc[i]}")
    print("-" * 50)
```

```python
print("\n=== FEATURE EXTRACTION AND TASK CLASSIFICATION ===")

# Prepare features for classification
# We'll classify tasks by category based on description and other features

# TF-IDF Vectorization
tfidf = TfidfVectorizer(max_features=1000, ngram_range=(1, 2))
X_text = tfidf.fit_transform(df['processed_description']).toarray()

# Additional numerical features
numerical_features = ['estimated_hours', 'user_workload']
X_numerical = df[numerical_features].values

# Encode categorical features
le_skill = LabelEncoder()
X_skill = le_skill.fit_transform(df['user_skill_level']).reshape(-1, 1)

# Combine all features
X_combined = np.hstack([X_text, X_numerical, X_skill])

# Target variable for classification (category)
le_category = LabelEncoder()
y_category = le_category.fit_transform(df['category'])

print(f"Feature matrix shape: {X_combined.shape}")
print(f"Classes: {le_category.classes_}")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y_category, test_size=0.2, random_state=42, stratify=y_category
)
```

```python

# TASK CLASSIFICATION MODELS


print("\n=== TRAINING TASK CLASSIFICATION MODELS ===")

# 1. Naive Bayes Classifier
print("Training Naive Bayes Classifier...")
nb_model = MultinomialNB(alpha=0.1)
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_pred)

print(f"Naive Bayes Accuracy: {nb_accuracy:.4f}")

# 2. SVM Classifier
print("Training SVM Classifier...")
svm_model = SVC(kernel='rbf', C=1.0, random_state=42)
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_pred)

print(f"SVM Accuracy: {svm_accuracy:.4f}")

# 3. Random Forest for comparison
print("Training Random Forest Classifier...")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"Random Forest Accuracy: {rf_accuracy:.4f}")

# Choose the best model
models = {'Naive Bayes': nb_accuracy, 'SVM': svm_accuracy, 'Random Forest': rf_accuracy}
best_model_name = max(models, key=models.get)
best_accuracy = models[best_model_name]

print(f"\nBest Classification Model: {best_model_name} with accuracy: {best_accuracy:.4f}")

# Detailed evaluation of the best model
if best_model_name == 'Naive Bayes':
    best_pred = nb_pred
elif best_model_name == 'SVM':
    best_pred = svm_pred
else:
    best_pred = rf_pred

print(f"\nDetailed Classification Report for {best_model_name}:")
print(classification_report(y_test, best_pred, target_names=le_category.classes_))
```

```python

print("\n=== PRIORITY PREDICTION MODEL ===")

# Prepare features for priority prediction
le_priority = LabelEncoder()
y_priority = le_priority.fit_transform(df['priority'])

# Use the same features as classification
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X_combined, y_priority, test_size=0.2, random_state=42, stratify=y_priority
)

# Random Forest for Priority Prediction
print("Training Random Forest for Priority Prediction...")
rf_priority = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
rf_priority.fit(X_train_p, y_train_p)
rf_priority_pred = rf_priority.predict(X_test_p)
rf_priority_accuracy = accuracy_score(y_test_p, rf_priority_pred)

print(f"Random Forest Priority Accuracy: {rf_priority_accuracy:.4f}")

# XGBoost for Priority Prediction
print("Training XGBoost for Priority Prediction...")
xgb_priority = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    random_state=42
)
xgb_priority.fit(X_train_p, y_train_p)
xgb_priority_pred = xgb_priority.predict(X_test_p)
xgb_priority_accuracy = accuracy_score(y_test_p, xgb_priority_pred)

print(f"XGBoost Priority Accuracy: {xgb_priority_accuracy:.4f}")

# Choose best priority model
priority_models = {'Random Forest': rf_priority_accuracy, 'XGBoost': xgb_priority_accuracy}
best_priority_model = max(priority_models, key=priority_models.get)
best_priority_accuracy = priority_models[best_priority_model]

print(f"\nBest Priority Model: {best_priority_model} with accuracy: {best_priority_accuracy:.4f}")
```

```python
print("\n=== HYPERPARAMETER TUNING ===")

# Grid Search for the best classification model
if best_model_name == 'Random Forest':
    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [10, 20],
        'min_samples_split': [2, 5]
    }
    grid_search = GridSearchCV(
        RandomForestClassifier(random_state=42),
        param_grid,
        cv=5,
        scoring='accuracy',
        n_jobs=-1
    )
    grid_search.fit(X_train, y_train)
    
    print(f"Best parameters for classification: {grid_search.best_params_}")
    print(f"Best CV score: {grid_search.best_score_:.4f}")
    
    # Update the model with best parameters
    best_classification_model = grid_search.best_estimator_
    tuned_pred = best_classification_model.predict(X_test)
    tuned_accuracy = accuracy_score(y_test, tuned_pred)
    print(f"Tuned model accuracy: {tuned_accuracy:.4f}")
else:
    best_classification_model = rf_model if best_model_name == 'Random Forest' else (nb_model if best_model_name == 'Naive Bayes' else svm_model)
    tuned_accuracy = best_accuracy
```

```python
print("\n=== WORKLOAD BALANCING SYSTEM ===")

class WorkloadBalancer:
    def __init__(self, df):
        self.df = df
        self.user_stats = self._calculate_user_stats()
    
    def _calculate_user_stats(self):
        user_stats = self.df.groupby('assigned_to').agg({
            'user_workload': 'mean',
            'user_skill_level': 'first',
            'actual_hours': 'sum',
            'task_id': 'count'
        }).rename(columns={'task_id': 'task_count'})
        
        return user_stats
    
    def recommend_assignment(self, task_priority, estimated_hours, required_skill):
        # Filter users by skill level
        skill_hierarchy = {'Beginner': 1, 'Intermediate': 2, 'Expert': 3}
        required_skill_level = skill_hierarchy.get(required_skill, 2)
        
        suitable_users = self.user_stats[
            self.user_stats['user_skill_level'].map(skill_hierarchy) >= required_skill_level
        ].copy()
        
        # Calculate workload score (lower is better)
        suitable_users['workload_score'] = (
            suitable_users['user_workload'] * 0.4 +
            suitable_users['actual_hours'] * 0.3 +
            suitable_users['task_count'] * 0.3
        )
        
        # Sort by workload score
        recommended_user = suitable_users.sort_values('workload_score').index[0]
        
        return recommended_user, suitable_users.loc[recommended_user, 'workload_score']

# Initialize workload balancer
balancer = WorkloadBalancer(df)

print("User Statistics:")
print(balancer.user_stats)

# Example recommendation
example_user, score = balancer.recommend_assignment('High', 5.0, 'Intermediate')
print(f"\nRecommended user for High priority, 5-hour task (Intermediate skill): {example_user}")
print(f"Workload score: {score:.2f}")
```

```python
print("\n=== AI TASK MANAGEMENT SYSTEM DEMO ===")

class IntelligentTaskManager:
    def __init__(self, classification_model, priority_model, workload_balancer, 
                 tfidf_vectorizer, label_encoders):
        self.classification_model = classification_model
        self.priority_model = priority_model
        self.workload_balancer = workload_balancer
        self.tfidf_vectorizer = tfidf_vectorizer
        self.le_category = label_encoders['category']
        self.le_priority = label_encoders['priority']
        self.le_skill = label_encoders['skill']
        self.nlp_processor = NLPPreprocessor()
    
    def process_new_task(self, task_description, estimated_hours, required_skill):
        # Preprocess description
        processed_desc = self.nlp_processor.preprocess_text(task_description)
        
        # Extract features
        text_features = self.tfidf_vectorizer.transform([processed_desc]).toarray()
        numerical_features = np.array([[estimated_hours, 5.0]])  # Default workload
        skill_encoded = self.le_skill.transform([required_skill]).reshape(-1, 1)
        
        # Combine features
        features = np.hstack([text_features, numerical_features, skill_encoded])
        
        # Predict category and priority
        predicted_category = self.le_category.inverse_transform([self.classification_model.predict(features)[0]])[0]
        predicted_priority = self.le_priority.inverse_transform([self.priority_model.predict(features)[0]])[0]
        
        # Recommend user assignment
        recommended_user, workload_score = self.workload_balancer.recommend_assignment(
            predicted_priority, estimated_hours, required_skill
        )
        
        return {
            'predicted_category': predicted_category,
            'predicted_priority': predicted_priority,
            'recommended_user': recommended_user,
            'workload_score': workload_score
        }

# Initialize the intelligent system
task_manager = IntelligentTaskManager(
    classification_model=best_classification_model,
    priority_model=xgb_priority if best_priority_model == 'XGBoost' else rf_priority,
    workload_balancer=balancer,
    tfidf_vectorizer=tfidf,
    label_encoders={
        'category': le_category,
        'priority': le_priority,
        'skill': le_skill
    }
)

# Demo with sample tasks
sample_tasks = [
    {
        'description': 'Fix critical security vulnerability in authentication system',
        'estimated_hours': 8.0,
        'required_skill': 'Expert'
    },
    {
        'description': 'Design new user interface for mobile application',
        'estimated_hours': 4.0,
        'required_skill': 'Intermediate'
    },
    {
        'description': 'Update documentation for API endpoints',
        'estimated_hours': 2.0,
        'required_skill': 'Beginner'
    }
]

print("\nProcessing sample tasks:")
print("=" * 80)

for i, task in enumerate(sample_tasks, 1):
    print(f"\nTask {i}: {task['description']}")
    print(f"Estimated Hours: {task['estimated_hours']}")
    print(f"Required Skill: {task['required_skill']}")
    print("-" * 50)
    
    result = task_manager.process_new_task(
        task['description'], 
        task['estimated_hours'], 
        task['required_skill']
    )
    
    print(f"Predicted Category: {result['predicted_category']}")
    print(f"Predicted Priority: {result['predicted_priority']}")
    print(f"Recommended User: {result['recommended_user']}")
    print(f"Workload Score: {result['workload_score']:.2f}")
```
